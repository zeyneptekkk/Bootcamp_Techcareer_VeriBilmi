```python
# Veri görselleştirme 
# y = x^2
import matplotlib.pyplot as plt
plt.style.use('ggplot')
x = np.arange(1,11) # [1 , 2, 3...,10 ]
y = x**2    #  [1, 4 ,9,..,100]
plt.plot(x,y)
plt.show()
```

```python
# sinüs ve cos grafiğini aynı yerde çiziniz.
x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8,4))
plt.plot(x,y1, label="Sin", color = "#79c2d0")
plt.plot(x,y2, label = 'Cos', color= "#ff6f3c")
plt.title('Sin(x) ve Cos(x)')
plt.xlabel('X')
plt.ylabel('Değerler')
plt.legend()
plt.show()
```

```python
# stok takip uygulaması 
data = {
    'Ürün adı' : ['Elma','Armut','Ay Çekirdeği','Şeker','Un','Yağ','Helva','Nohut Dürümü','Kayısı','Fıstık','Lokum'],
    'Stok': np.random.randint(0,200,11),
    'Fiyat': np.random.randint(25,150,11),
    'Kategori': ['Meyve','Meyve','Çerez','Erzak','Erzak','Erzak','Atıştırmalık','Yemek','Meyve','Çerez','Çerez']
}
```

```python
df1 = pd.DataFrame(data)
df1
```

```python
# Kdv siz fiyatı verilmiştir. kdv li fiyatı da ayrı bir kolon ile gösterilmesi istenilmektedir.
df1['Kdv li fiyat'] = df1['Fiyat'] * 1.2
df1
```

```python
# KDV li fiyattan tüm ürünleri sattığımızı varsayalım. her ürün için ne kadar tl lik gelir elde ettiğimizi ayrı kolon da gösteriniz.
df1['Gelir'] = df1['Stok'] * df1['Kdv li fiyat']
df1
```

```python
# Kategoriye göre Toplam gelirleri gruplayan listeyi veriniz.
gruplanmis_kategori = df1.groupby('Kategori')['Gelir'].sum()
```

```python
# Gruplanmış Ürünleri bar grafikte gösteriniz.
plt.bar(gruplanmis_kategori.keys(), gruplanmis_kategori.values)
plt.title('Kategoriye göre toplam gelir')
plt.show()
```

```python
plt.pie(gruplanmis_kategori.values, labels= gruplanmis_kategori.keys(), autopct=lambda p : f'%{p:.2f} \n {p * sum(gruplanmis_kategori.values)/100:.0f}')
plt.show()
```
