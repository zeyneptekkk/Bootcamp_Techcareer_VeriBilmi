```python

daf1 = pd.DataFrame({'id':[1,2,3,4,5], 'Name':['Ömer','Buse','Akça','Sema','Yusuf']})
daf1
```

```python

daf2 = pd.DataFrame({
    'id' : [1,2,3,4,5,6,7],
    'Takım': ['Malatyaspor','Ümraniyespor','Galatasaray','Fenerbahçe','Beşiktaş','Trabzonspor','Başakşehir']
})
daf2
```

```python
# inner join
pd.merge(daf1,daf2,on="id",how="inner")
```

```python
# outer join
pd.merge(daf1,daf2,on="id",how="outer")
```
